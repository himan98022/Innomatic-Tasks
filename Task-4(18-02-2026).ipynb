```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Recharge must be at least ₹50")
        elif amount not in valid_plans:
            print("Invalid plan! Try again.")
        else:
            print("Recharge Successful ✅")
            break

validate_recharge()
```

```python
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, ": Reorder Alert ⚠")
        else:
            print(item, ": Stock OK")

products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Salt": 30
}

check_inventory(products)
```

```python
def student_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 45, 70, 55]
result = student_result(marks)

print("Result:", result)
```

```python
def calculate_fare(km, peak):
    fare = 50 + (12 * km)

    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare


while True:
    km = float(input("Enter distance (km): "))
    peak = input("Peak hour? (yes/no): ")

    total = calculate_fare(km, peak)
    print("Total Fare = ₹", total)

    retry = input("Retry? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
def attendance_check(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

record = ["P", "P", "A", "P", "P", "A", "P", "P"]
print(attendance_check(record))
```

```python
def check_password(password):
    special = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


pwd = input("Enter password: ")
print(check_password(pwd))
```
