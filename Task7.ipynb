```python
# Class for Smart Light Controller
class SmartLight:

    # Constructor to initialize light name and status
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # default status

    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"

    # Method to display status
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Example usage
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

```python
# Class to store employee details
class EmployeeIDCard:

    # Constructor
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Example usage
employee = EmployeeIDCard("Rahul", "EMP102", "IT")
employee.display_id_card()
```

```python
# Class for storing contact details
class Contact:

    # Constructor
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact info
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Example usage
contact = Contact("Anita", "9876543210")
contact.display_contact()
```

```python
# Class for product price tag
class Product:

    # Constructor
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    # Method to display price tag
    def display_price_tag(self):
        print("Product:", self.product_name)
        print("Price: ₹", self.price)


# Example usage
product = Product("Headphones", 2499)
product.display_price_tag()
```

```python
# Class for movie rating
class Movie:

    # Constructor
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    # Method to display movie details
    def display_movie(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")


# Example usage
movie = Movie("Inception", 4.8)
movie.display_movie()
```

```python
# Class for delivery details
class Delivery:

    # Constructor
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to display delivery details
    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Example usage
delivery = Delivery("Suman", "Hyderabad")
delivery.display_delivery()
```
