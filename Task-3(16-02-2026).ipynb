```python
# Problem 1: Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

# Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers),
      f"(Score: {highest_score})")
```

```python
# Problem 2: Search Query Keyword Analysis

import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Split words
words = query.split()

# Count frequency
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Display words occurring more than once
result = {word: count for word, count in word_count.items() if count > 1}

print(result)
```

```python
# Problem 3: Sensor Data Validation

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # even number check
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem 4: Email Domain Usage Analysis

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Calculate percentage
total = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem 5: Sales Spike Detection

# Input sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average daily sales
average_sales = sum(sales) / len(sales)
print("Average sales:",average_sales);

# Step 2: Calculate spike threshold (30% above average)
threshold = average_sales * 1.30

# Step 3: Detect and display spike days
for i, sale in enumerate(sales):
    if sale > threshold:
        print(f"Day {i+1}: {sale}")
```

```python
# Problem 6: Duplicate User ID Detection

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count occurrences
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

# Display duplicates
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
